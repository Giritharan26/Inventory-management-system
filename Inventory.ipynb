# Load the products to E-wall

```python
fd=open("products.txt",'a')
no=int(input("How many Product you want add it: "))
for i in range(no):
    data0=input("Enter the Product_id: " )
    data1=input("Enter the name: " )
    data2=input("Enter the price: " )
    data3=input("Enter the quantity: ")

    data=data0+','+data1+','+data2+','+data3+'\n'

    fd.write(data)
    print('\n')

#print(data)
fd.close()
```

# Buy a single product

```python
fd=open("products.txt",'r')
#print(fd.read())

data= fd.read()
data1=data.split('\n')
print("************* List of Products *************\n")
print("----------------------------------------")
print("Product Id      Name    Price   Quantity")
print("----------------------------------------")
for i in data1:
    j=i.split(',')
    print(j[0]+'\t\t'+j[1]+'\t'+j[2]+'\t'+j[3]+'\n')
print("----------------------------------------")
fd.close()

print("*** Which product You want to buy ***")

p_id = input("Enter the product id: ")
p_qu = input("Enter the no of Quantity: ")

fd=open("products.txt",'r')
pro=fd.read().split('\n')
#print(type(pro))
for i in pro:
    i=i.split(',')
    if i[0]==p_id:
        print("------------------------------------------------")
        print("Product Id      Name    Price   Quantity   Total")
        print("------------------------------------------------")
        print(i[0]+'\t\t'+i[1]+'\t'+i[2]+'\t'+p_qu+'\t   '+str(int(i[2])*int(p_qu))+'\n')
        print("------------------------------------------------")
        print("Bill Amount: \t\t\t\t   "+str(int(i[2])*int(p_qu)))
        print("------------------------------------------------")

fd.close()
```

```python
fd=open("products.txt",'r')
#print(fd.read())

data= fd.read()
data1=data.split('\n')
print("************* List of Products *************\n")
print("----------------------------------------")
print("Product Id      Name    Price   Quantity")
print("----------------------------------------")
for i in data1:
    j=i.split(',')
    print(j[0]+'\t\t'+j[1]+'\t'+j[2]+'\t'+j[3]+'\n')
print("----------------------------------------")
fd.close()

def select_pro(p_id, p_qu, pro):
    for i in pro:
        i=i.split(',')
        if i[0]==p_id:
            tat=str(int(i[2])*int(p_qu))
            print(i[0]+'\t\t'+i[1]+'\t'+i[2]+'\t'+p_qu+'\t   '+tat+'\n')
            return tat

print("*** Which product You want to buy ***")

fd=open("products.txt",'r')
pro=fd.read().split('\n')
#print(type(pro))

p_no = input("Enter the no of product you to buy: ")
sum=0
j=0
for i in range(int(p_no)):
    if j==0:
        print("------------------------------------------------")
        print("Product Id      Name    Price   Quantity   Total")
        print("------------------------------------------------")
        j=j+1
    p_id = input("Enter the product id: ")
    p_qu = input("Enter the no of Quantity: ")
    sum+=int(select_pro(p_id, p_qu, pro))



#go to select_product function
print("------------------------------------------------")
print("Bill Amount: \t\t\t\t   "+str(sum))
print("------------------------------------------------")

fd.close()
```
